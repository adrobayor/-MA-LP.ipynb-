# Nyquist and Aliasing

```julia
using PyPlot
```

```julia
frec=45 #Frecuencia característica
sps=100 #samples per second
seg=5 #tiempo de muestreo
t=collect(linspace(0,seg,seg*sps)); #Vector de tiempo
```

```julia
y=sin.(2*π*frec*t);
```

```julia
plot(t,y)
```

```julia
f=abs.(fft(y,1));
```

```julia
frecvec=collect(linspace(0,sps,sps*seg))
length(frecvec)
```

```julia
plot(frecvec[1:250],f[1:250])
```
